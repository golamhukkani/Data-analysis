```python
highest_sales = df.groupby(['Name'], as_index=False)['Global_Sales'].sum().sort_values(by='Global_Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(30,8)})
sns.barplot(data = highest_sales , x = 'Name',y= 'Global_Sales')
```
