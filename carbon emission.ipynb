```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('D:\carbon emissions.csv',encoding = 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
highest_emission = df.groupby(['Country'], as_index=False)['Total'].sum().sort_values(by='Total', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_emission , x = 'Country',y= 'Total')
```

U.S.A and China has the highest carbon emission

```python
highest_emission_year = df.groupby(['Year'], as_index=False)['Total'].sum().sort_values(by='Total', ascending=False).head(30)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_emission_year , x = 'Year',y= 'Total')
```

The carbon emission is increasing every year for the last 30 years.

```python
highest_oil_year = df.groupby(['Year'], as_index=False)['Oil'].sum().sort_values(by='Oil', ascending=False).head(30)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_oil_year , x = 'Year',y= 'Oil')
```

Oil usage is increasing at a slow rate.

```python
highest_gas_year = df.groupby(['Year'], as_index=False)['Gas'].sum().sort_values(by='Gas', ascending=False).head(30)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_gas_year , x = 'Year',y= 'Gas')
```

Gas usage is increasing rapidly every year.

```python
highest_cement_year = df.groupby(['Year'], as_index=False)['Cement'].sum().sort_values(by='Cement', ascending=False).head(30)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_cement_year , x = 'Year',y= 'Cement')
```

Cement usage is also increasing very fast every year.

```python
highest_coal_year = df.groupby(['Year'], as_index=False)['Coal'].sum().sort_values(by='Coal', ascending=False).head(30)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_coal_year , x = 'Year',y= 'Coal')
```

Coal usage is decreasing slowly in the recent years.

```python
highest_flaring_year = df.groupby(['Year'], as_index=False)['Flaring'].sum().sort_values(by='Flaring', ascending=False).head(30)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_flaring_year , x = 'Year',y= 'Flaring')
```

Flaring emissions increased slowly and decreasing in the recent years.

```python
highest_oil_country = df.groupby(['Country'], as_index=False)['Oil'].sum().sort_values(by='Oil', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_oil_country , x = 'Country',y= 'Oil')
```

U.S.A has the highest carbon emission from oil.

```python
highest_cement_country = df.groupby(['Country'], as_index=False)['Cement'].sum().sort_values(by='Cement', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_cement_country , x = 'Country',y= 'Cement')
```

China has the highest emission from cement.

```python
highest_gas_country = df.groupby(['Country'], as_index=False)['Gas'].sum().sort_values(by='Gas', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_gas_country , x = 'Country',y= 'Gas')
```

U.S.A has the highest emission from gas and Russia is second.

```python
highest_coal_country = df.groupby(['Country'], as_index=False)['Coal'].sum().sort_values(by='Coal', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_coal_country , x = 'Country',y= 'Coal')
```

China has the highest emission from coal and then U.S.A.

```python
highest_percapita_country = df.groupby(['Country'], as_index=False)['Per Capita'].sum().sort_values(by='Per Capita', ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = highest_percapita_country , x = 'Country',y= 'Per Capita')
```

U.S.A has the highest per capita carbon emission and then Luxemburg and at third Australia.
